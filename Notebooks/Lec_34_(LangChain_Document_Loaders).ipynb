# <span style='background :lightgreen' >1. Understanding LangChain **Indexes** Component</span>

<h3 align="center"><div class="alert alert-success" color=magenta style="margin: 20px">LangChain Indexes consist of four interconnected components that work together to enable Retrieval Augmented Generation (RAG).</h3>

<img align="right" width="800" src="../images/r3.png"  >

- [**Document Loaders**](https://docs.langchain.com/oss/javascript/integrations/document_loaders) provide a standard interface (160+ integrations) for reading data from different sources ike PDFs, web pages, Google Drive, Slack, Notion, and more into LangChain’s Document format. This ensures that data can be handled consistently regardless of the source. Common categories include:
    - **TextLoaders**  for plain .txt files (`CSVLoader`, `Docx2txtLoader`, `JSONLoader`, `PyPDFLoader`, `MarkdownLoader` etc)
    - **Web loaders** for ingesting data from online or cloud-based sources. (`WebBaseLoader`, `SitemapLoader`, `S3FileLoader`, `GCSFileLoader`, etc)
    - **Productivity tools** for pulling data from collaboration and enterprise platforms. (`GoogleDriveLoader`, `YouTubeLoader`, `TwitterLoader`, `WikipediaLoader`, `NotionPageLoader`, `ConfluenceLoader` etc)
- [**Text Splitters**](https://docs.langchain.com/oss/javascript/integrations/splitters/index#text-splitters) break large docs into smaller chunks that will be retrievable individually, ensuring that each chunk contains meaningful information while staying within model context window limit. Main strategies are:
    - **Text structure-based:** Splits hierarchically (paragraphs → sentences → words)
    - **Length-based:** Splits by token or character count
    - **Document structure-based:** Splits based on format (HTML, Markdown, code)
- [**Vector Stores**](https://docs.langchain.com/oss/python/integrations/vectorstores/index) are specialized databases for storing and searching embeddings. They store embedded documents and perform similarity search to find semantically similar content. LangChain provides a unified interface with methods like addDocuments, delete, and similaritySearch. Popular options include Chroma, Pinecone, Qdrant, FAISS, MongoDB Atlas, and in-memory stores for testing.
- [**Retrievers**](https://docs.langchain.com/oss/python/integrations/retrievers/index#retrievers) accepts an unstructured query as input, search through the vector database to find the most relevant document chunks and return a list of Documents as output.

# <span style='background :lightgreen' >4. Document Loaders in LangChain</span>

<h3 align="center"><div class="alert alert-success" color=magenta style="margin: 20px">Document loaders are specialized tools that connect to different data sources and extract text content from various file formats. Think of them as translators that convert different types of documents into a standard text format that the RAG system can understand and process.</h3>

<h3 align="center"><div class="alert alert-success" color=magenta style="margin: 20px">Once Document loaders extract and standardize the content, the text flows to Text Splitters for chunking, then to embedding models for vector conversion, and finally to vector databases for storage and retrieval.</h3>



- **How Document Loaders Work:** Document loaders follow a simple process:
    1. **Connect** to the data source (file system, web, database)
    2. **Read** the document in its original format
    3. **Extract** the text content while preserving important information
    4. **Convert** the content into a standardized format for further processing
    5. **Pass** the extracted text to the next component (Text Splitters)

> Visit for different Document Loaders: https://python.langchain.com/docs/integrations/document_loaders/

## b. Example: `TextLoader`
<h3 align="center"><div class="alert alert-success" color=magenta style="margin: 20px">TextLoader is the most simple and commonly used Document loader in LangChain that reads plain text (.txt) files and convert them into LangChain Document objects.</h3>

- **Purpose**: Handles plain text files (.txt, .md, .log)  
- **Use Cases**: Loading log files, scraped text, video transcripts, markdown files, code snippets and configuration files  
- **Advantages**: Fast and straightforward, no complex parsing needed  
- **Example**: Loading README files, documentation, or plain text data

````python
# Import the appropriate loader
from langchain_community.document_loaders import TextLoader

# Create loader instance by passing it the file  path, which will load the .txt file in a LangChain Document object
loader = TextLoader('../data/arif_bio.txt', encoding='utf-8')

# Load the .txt file and returns a list of Documents with each Document having two attributes
# page_content: which is a string containing actual text content
# metadata: which is a dictionary containing information about the document (like source file path)
docs = loader.load()

print(type(docs))                      # Shows the data type        
print(f"Loaded {len(docs)} document")  # Text loaders consider whole text file as one chunk  
print(f"\nMetadata: {docs[0].metadata}")        # Crucial for filtering search results, tracking document sources, debugging and auditing.
print(f"File Content: {docs[0].page_content[:200]}...")      # Will be passed to text splitter
````

## d. Example `PyPDFLoader`

<h3 align="center"><div class="alert alert-success" color=magenta style="margin: 20px">PyPDFLoader is a Document loader in LangChain used to load content from a pdf file and convert each page into a LangChain Document objects.</h3>

- **Purpose**: Extracts text from PDF documents. 
- **Use Cases**: Academic papers, reports, manuals, and business documents  
- **Advantages**: Preserves page information,. Every page of Pdf load as one document  
- **Example**: Loading research papers, user manuals, or financial reports for analysis
- **PyPDFLoader Limitations:** PyPDFLoader uses the PyPDF library under the hood, which has some limitations:
    - **Not great with scanned PDFs** - Struggles with image-based PDFs that don't contain actual text
    - **Complex layouts** - May have difficulty with PDFs containing tables, columns, or complex formatting
    - **Structural issues** - Can encounter problems with PDFs that have internal structural problems (as we saw with the warning messages)
- **Choosing the Right PDF Loader:**

| **Use Case** | **Recommended Loader** |
|-------------|----------------------|
| Simple, clean PDFs | `PyPDFLoader` |
| Fast processing and can extract images as well | `PyMuPDFLoader` |
| PDFs with tables/columns | `PDFPlumberLoader` |
| Scanned/image PDFs | `UnstructuredPDFLoader` or `AmazonTextractPDFLoader` |
| Want best structure extraction | `UnstructuredPDFLoader` |

- **PDF Challenges:** PDFs are notoriously difficult to parse because they:
    - Store text in complex ways (not just simple text)
    - Can have formatting issues
    - May contain scanned images (requiring OCR)
    - Often have extraction artifacts

> For details visit: https://python.langchain.com/docs/how_to/document_loader_pdf/

## e. Example `CSVLoader` and `UnstructuredCSVLoader`

## h. Example `DirectoryLoader`
<h3 align="center"><div class="alert alert-success" style="margin: 20px">DirectoryLoader is a Document loader in LangChain used to load content from all files within a specified directory and convert each file into LangChain Document objects based on the configured file type loader.</div></h3>

- **DirectoryLoader** is a document loader that lets you load multiple documents from a directory (folder) of files. A folder with `report.pdf`, `notes.txt`, and `data.csv` will all be loaded as documents in a single operation.
- DirectoryLoader automatically detects and loads different file formats:
    - **PDFs** → Uses PyPDFLoader
    - **Text files** → Uses TextLoader  
    - **CSV files** → Uses CSVLoader
    - **Word docs** → Uses Docx2txtLoader
    - And more...

| Glob Pattern | What It Loads |
|--------------|---------------|
| `**/*.txt` | All `.txt` files in all subfolders |
| `*.pdf` | All `.pdf` files in the root directory |
| `data/*.csv` | All `.csv` files in the `data/` folder |
| `**/*` | All files (any type, all folders) |

> **Note:** `**` = recursive search through subfolders

**Characteristics of `DirectoryLoader` class:**
- ✅ Advantages:
    - Loads multiple files at once
    - Supports glob patterns
    - Progress tracking
    - Recursive directory scanning
- ❌ Disadvantages:
    - All files must be same type
    - Limited error handling per file
    - Can be memory intensive for large directories
- **`.load()` performs Eager Loading**, i.e., loads all documents immediately into memory and returns a list of Document objects. Use it when the number of documents is small and you want everything loaded upfront. It can have following problems with large documents:
    - **Memory overload:** All documents stored in RAM at once
    - **Slow startup:**  Must wait for all files to load before processing
    - **System crashes:** Can run out of memory with many/large files
- **`.lazy_load()` loads on demand**, i.e., the documents are not loaded at once. It returns a generator of document objects and you can fetch one at a time as needed. It hs following advantages:
    - **Memory Management:**  It only loads one document at a time into memory, then discards it after processing. With `load()`, if you have 100 large PDFs, all 100 are stored in RAM simultaneously, which can crash your system.
    - **Streaming Processing:**  You can start processing immediately without waiting for all files to load. It follows a pattern: load document → process/chunk → embed → discard → load next document.
    - **Handles unlimited files:**  Can process thousands of documents without memory issues

````python
# Import the appropriate loader
from langchain_community.document_loaders import DirectoryLoader, TextLoader

# load all the text files from the directory
loader = DirectoryLoader(
                        "../data/",
                        glob="**/*.txt", ## Pattern to match files  
                        loader_cls= TextLoader, ##loader class to use
                        loader_kwargs={'encoding': 'utf-8'},
                        show_progress=True
                    )

docs = loader.load()  

print(f"Loaded {len(docs)} documents")
for i, doc in enumerate(docs):
    print(f"\nDocument {i+1}:")
    print(f"  Metadata: {doc.metadata}")
    print(f"  Length: {len(doc.page_content)} characters")
    print(f"  File Content: {doc.page_content[:100]}...")  
````

## i. Example `SQLDatabaseLoader`

<h3 align="center"><div class="alert alert-success" style="margin: 20px">SQLDatabaseLoader is a Document loader in LangChain used to load content from SQL database tables by executing queries and converting each query result row into a LangChain Document object.</div></h3>

>- SQLite is a serverless dbms used with mobile apps, desktop apps, embedded systems and the database file moves with the app.
>- You can access the db using the file path of your OS with OS permissions, so there is no user management. 

**Database Schema:**
1. **Two related tables**: `programs` and `students`
2. **Foreign Key**: `students.program_id` references `programs.program_id`
3. **Real relationships**: Each student is enrolled in a specific program at a university

```
┌─────────────────────────┐
│      programs           │
├─────────────────────────┤
│ program_id (PK)         │
│ program_name            │
│ university              │
│ duration_years          │
└─────────────────────────┘
           ↑
           │ (Foreign Key)
           │
┌─────────────────────────┐
│      students           │
├─────────────────────────┤
│ id (PK)                 │
│ name                    │
│ age                     │
│ program_id (FK)         │
│ enrollment_year         │
└─────────────────────────┘
```

````python
import sqlite3
import os

# Delete the database file if it exists
if os.path.exists("../data/temp_students.db"):
    os.remove("../data/temp_students.db")
    print("Old database deleted.\n")

# Create a new database
conn = sqlite3.connect("../data/temp_students.db")
cursor = conn.cursor()

# Create TABLE 1: programs containing Degree Programs and university
cursor.execute('''
    CREATE TABLE IF NOT EXISTS programs (
        program_id INTEGER PRIMARY KEY AUTOINCREMENT,  -- Unique ID for each program
        program_name TEXT NOT NULL,                    -- Name of the degree program
        university TEXT NOT NULL,                      -- University offering the program
        duration_years INTEGER                         -- Program duration in years
    )
''')
# Insert sample programs
programs = [
    ('Computer Science', 'MIT', 4),
    ('Data Science', 'Stanford University', 2),
    ('Business Administration', 'Harvard University', 2),
    ('Electrical Engineering', 'MIT', 4),
    ('Medicine', 'Johns Hopkins University', 6)
]
cursor.executemany('INSERT INTO programs (program_name, university, duration_years) VALUES (?, ?, ?)', programs)
print("Programs table created and populated\n")


# Create TABLE 2: students containing name, age (with foreign key to programs)
cursor.execute('''
    CREATE TABLE IF NOT EXISTS students (
        id INTEGER PRIMARY KEY AUTOINCREMENT,          -- Auto-incrementing student ID
        name TEXT NOT NULL,                            -- Student name
        age INTEGER,                                   -- Student age
        program_id INTEGER,                            -- Foreign key linking to programs table
        enrollment_year INTEGER,                       -- Year student enrolled
        
        -- Foreign key constraint: program_id must exist in programs table
        FOREIGN KEY (program_id) REFERENCES programs(program_id)
    )
''')
# Insert sample students (now with program_id and enrollment_year)
students = [
    ('Alisha', 21, 1, 2025),    # Enrolled in Computer Science at MIT
    ('Abeeha', 22, 2, 2022),    # Enrolled in Data Science at Stanford
    ('Khalood', 22, 2, 2022),   # Enrolled in Data Science at Stanford
    ('Jamil', 30, 3, 2024),     # Enrolled in Business Administration at Harvard
    ('Rauf', 42, 4, 2023),      # Enrolled in Electrical Engineering at MIT
    ('Haider', 36, 5, 2025),    # Enrolled in Medicine at Johns Hopkins
    ('Mujahid', 25, 1, 2023)    # Enrolled in Computer Science at MIT
]
cursor.executemany('INSERT INTO students (name, age, program_id, enrollment_year) VALUES (?, ?, ?, ?)', students)
print("Students table created and populated\n")

print("Database created successfully with two related tables!")

# Show all programs
print("PROGRAMS TABLE:")
print("-" * 80)
cursor.execute("SELECT * FROM programs")
programs_results = cursor.fetchall()
for row in programs_results:
    print(f"   ID: {row[0]} | Program: {row[1]} | University: {row[2]} | Duration: {row[3]} years")

# Show all students
print("\nSTUDENTS TABLE:")
print("-" * 80)
cursor.execute("SELECT * FROM students")
students_results = cursor.fetchall()
for row in students_results:
    print(f"   ID: {row[0]} | Name: {row[1]} | Age: {row[2]} | Program ID: {row[3]} | Enrolled: {row[4]}")

# Show students with their program details (JOIN query)
print("\nSTUDENTS WITH PROGRAM DETAILS (JOIN):")
print("-" * 80)
cursor.execute("""
    SELECT 
        s.id,
        s.name,
        s.age,
        p.program_name,
        p.university,
        s.enrollment_year
    FROM students s
    JOIN programs p ON s.program_id = p.program_id
    ORDER BY s.name
""")
joined_results = cursor.fetchall()
for row in joined_results:
    print(f"   {row[1]} (Age {row[2]}) → {row[3]} at {row[4]} (Enrolled: {row[5]})")

# Commit and close
conn.commit()
conn.close()
````

````python
# Custom SQL to Document conversion for students database with relationships
import sqlite3
from langchain_core.documents import Document
from typing import List

def sql_to_documents(db_path: str) -> List[Document]:
    """
    Convert SQL Database to LangChain documents with context.
    
    This function reads a SQLite database and creates Document objects
    that can be used for RAG (Retrieval Augmented Generation).
    It handles both individual tables and their relationships.
    
    Args:
        db_path: Path to the SQLite database file
        
    Returns:
        List of Document objects containing table information and relationships
    """
    
    # Step 1: Connect to the SQLite database
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    
    # Initialize empty list to store all documents
    documents = []
    
    # Step 2: Get list of all tables in the database
    # sqlite_master is a special table that stores database schema
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tables = cursor.fetchall()  # Returns list of tuples: [('students',), ('programs',)]
    
    print(f"Found {len(tables)} table(s) in database\n")
    
    # Step 3: Process each table individually
    for table in tables:
        table_name = table[0]  # Extract table name from tuple
        print(f"Processing table: {table_name}")
        
        # Step 4: Get column information (schema) for this table
        # PRAGMA table_info returns: (column_id, column_name, data_type, not_null, default_value, primary_key)
        cursor.execute(f"PRAGMA table_info({table_name});")
        columns = cursor.fetchall()
        
        # Extract just the column names (index 1 in each tuple)
        column_names = [col[1] for col in columns]
        print(f"   Columns: {column_names}")
        
        # Step 5: Fetch all data from this table
        cursor.execute(f"SELECT * FROM {table_name}")
        rows = cursor.fetchall()  # Returns list of tuples, one per row
        print(f"   Records: {len(rows)}\n")
        
        # Step 6: Build the document content as a formatted string
        # This will be the actual text that gets embedded and searched
        table_content = f"Table: {table_name}\n"
        table_content += f"Columns: {', '.join(column_names)}\n"
        table_content += f"Total Records: {len(rows)}\n\n"
        
        # Step 7: Add all records (or sample if too many)
        table_content += "Records:\n"
        for row in rows:  # Show all records (for small tables)
            # Convert row tuple to dictionary for better readability
            # zip() pairs column names with values: {'id': 1, 'name': 'Alisha', 'age': 21}
            record = dict(zip(column_names, row))
            table_content += f"{record}\n"
        
        # Step 8: Create a LangChain Document object
        # page_content: The actual text that will be embedded and searched
        # metadata: Additional information for filtering and context
        doc = Document(
            page_content=table_content,
            metadata={
                'source': db_path,              # Where this data came from
                'table_name': table_name,       # Which table
                'num_records': len(rows),       # How many records
                'columns': column_names,        # What columns exist
                'data_type': 'sql_table'        # Type of data
            }
        )
        
        # Add this document to our collection
        documents.append(doc)
    
    # ========================================================================
    # Step 9: Create a relationship document by JOINing tables
    # ========================================================================
    print("🔗 Creating relationship document (JOIN)...\n")
    
    # Execute JOIN query to combine students with their program details
    cursor.execute("""
        SELECT 
            s.id,
            s.name,
            s.age,
            s.enrollment_year,
            p.program_name,
            p.university,
            p.duration_years
        FROM students s
        JOIN programs p ON s.program_id = p.program_id
        ORDER BY s.name
    """)
    
    # Fetch all joined results
    relationships = cursor.fetchall()
    
    # Build the relationship document content
    rel_content = "Student-Program Relationships:\n\n"
    rel_content += "This document shows which students are enrolled in which programs and universities.\n\n"
    
    # Add each relationship in a human-readable format
    for rel in relationships:
        student_id, name, age, enrollment_year, program, university, duration = rel
        rel_content += f"• {name} (ID: {student_id}, Age: {age}) is enrolled in {program} "
        rel_content += f"at {university}. Enrolled in {enrollment_year}. "
        rel_content += f"Program duration: {duration} years.\n"
    
    # Create the relationship document
    rel_doc = Document(
        page_content=rel_content,
        metadata={
            'source': db_path,
            'data_type': 'sql_relationships',
            'relationship': 'student_program_join',
            'num_relationships': len(relationships),
            'tables_involved': ['students', 'programs']
        }
    )
    
    # Add relationship document to our collection
    documents.append(rel_doc)
    print(f"✅ Created relationship document with {len(relationships)} connections\n")
    
    # Step 10: Close database connection
    conn.close()
    print(f"✅ Created {len(documents)} document(s) total from database\n")
    
    # Return all documents (individual tables + relationships)
    return documents


# ============================================================================
# Test the function with our students database
# ============================================================================
print("=" * 80)
print("Converting students.db to Documents")
print("=" * 80 + "\n")

docs = sql_to_documents("../data/temp_students.db")

# Display the created documents
print("=" * 80)
print("Generated Documents:")
print("=" * 80 + "\n")

for i, doc in enumerate(docs, 1):
    print(f"Document {i}:")
    print(f"Content:\n{doc.page_content}\n")
    print(f"Metadata: {doc.metadata}\n")
    print("-" * 80 + "\n")
````
