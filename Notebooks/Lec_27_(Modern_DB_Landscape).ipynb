## a. Comparison of SQL, NoSQL, and Vector Databases


| Feature | SQL Database (Relational) | NoSQL Database (Non-Relational) | Vector Database (AI-Native) |
|---------|---------------------------|----------------------------------|------------------------------|
| **Data Structure** | Structured tables with predefined schemas (rows and columns) | Flexible: documents, key-value pairs, graphs, or wide-column stores | High-dimensional vectors (embeddings) with optional metadata |
| **Data Type** | Text, numbers, dates, booleans stored in fixed columns | JSON/BSON documents, nested objects, arrays, varied data types | Numerical arrays (vectors) representing semantic meaning of data |
| **Schema** | Strict schema required before inserting data | Schema-less or dynamic schema (flexible structure) | Minimal schema: vectors + optional metadata (schema-flexible) |
| **Relationships** | Strong relationships using foreign keys and JOINs | Embedded documents or references (denormalization preferred) | Weak relationships; focuses on content similarity rather than explicit links |
| **Scalability** | Vertical scaling (scale up); horizontal scaling is complex | Horizontal scaling (scale out); designed for distributed systems | Horizontal scaling; optimized for high-dimensional vector operations |
| **ACID Compliance** | Strong ACID guarantees (Atomicity, Consistency, Isolation, Durability) | Often eventual consistency; some offer ACID for single documents | Varies; typically prioritizes performance over strict ACID |
| **Query Language** | SQL (Structured Query Language) | Varies by database: MongoDB Query Language, CQL, Cypher, etc. | Similarity search APIs, distance metrics (cosine, euclidean, dot product) |
| **Query Style** | • Exact match queries<br>• Filtering, sorting, joining<br>• Example: `WHERE name = 'Ali'` | • Document/field queries<br>• Range queries, text search<br>• Example: `{name: "Ali", age: {$gt: 25}}` | • Semantic similarity search<br>• Nearest neighbor search<br>• Example: "Find documents similar to 'machine learning'" |
| **Query Example** | `SELECT * FROM users WHERE age > 25 AND city = 'Lahore'` | `db.users.find({age: {$gt: 25}, city: "Lahore"})` | `collection.query(query_texts=["young professionals in Lahore"])` |
| **Primary Use Cases** | • Financial transactions<br>• E-commerce systems<br>• Inventory management<br>• Banking applications<br>• HR systems | • Content management<br>• Real-time analytics<br>• Social media platforms<br>• IoT data storage<br>• Mobile applications | • Semantic search engines<br>• AI chatbots & RAG systems<br>• Recommendation systems<br>• Image/video similarity search<br>• Anomaly detection |
| **Popular Examples** | • MySQL<br>• PostgreSQL<br>• Oracle Database<br>• Microsoft SQL Server<br>• SQLite<br>• MariaDB | **Document:** MongoDB, CouchDB<br>**Key-Value:** Redis, DynamoDB<br>**Column:** Cassandra, HBase<br>**Graph:** Neo4j, ArangoDB | • ChromaDB<br>• Pinecone<br>• Weaviate<br>• Milvus<br>• Qdrant<br>• FAISS (by Meta) |


#### Use SQL Database When:
- Data has clear relationships and structure
- You need ACID compliance and data integrity
- Complex queries with JOINs are required
- Data consistency is critical
#### Use NoSQL Database When:
- Data structure is flexible or evolving
- You need to scale horizontally
- High read/write throughput is required
- Working with semi-structured or unstructured data
#### Use Vector Database When:
- Building AI/ML applications
- Need semantic search capabilities
- Working with embeddings from text, images, or audio
- Implementing recommendation systems or similarity search

## e. CAP Theorem
- The CAP theorem was introduced by computer scientist Eric Brewer in 2000
- CAP stands for the three key properties of a distributed system out of which you can only guarantee <strong>two out of three </strong> properties in any distributed system.
    - **Consistency:**  All nodes see the same data
    - **Availability:**  Every request gets a response
    - **Partition Tolerance:**  Can handle network failure

# <span style='background :lightgreen' >2. Why One Database Isn’t Enough Anymore (Polyglot Persistence)</span>

<h3 align="center"><div class="alert alert-success" color=magenta style="margin: 20px">An e-commerce site might use PostgreSQL for orders, MongoDB for product catalogs, and ChromaDB for product recommendations</h3>
    
- **Polyglot Persistence:** is an architectural approach where systems use multiple specialized databases, each optimized for specific tasks. Rather than forcing all data into a single database technology, modern applications strategically combine different database types to maximize performance, scalability, and functionality.
- **Why Use Multiple Databases?** Modern systems have diverse requirements that no single database can optimally satisfy:
    - Performance optimization – Different operations require different storage mechanisms
    - Scalability patterns – Various data types scale differently
    - Cost efficiency – Use expensive resources only where truly needed
    - Flexibility – Adapt to changing requirements without complete rewrites
    
## a. Traditional ERP or Banking Systems
- What they store:
    - Customer accounts and records
    - Financial transactions
    - Product catalogs
    - Complex business relationships
- What they need:
    - ACID compliance for data integrity
    - Strong relational integrity and foreign keys
    - Robust, well-defined schemas
    - Complex join operations and transactions
    - Proven reliability and maturity
- Best database choices:
    - PostgreSQL – Advanced features, excellent for complex queries
    - MySQL – Widely adopted, reliable, great ecosystem


## b. Real-Time Mobile or Web Applications
- What they store:
    - User profiles and preferences
    - Session data and application state
    - User-generated content
    - Settings and configurations
- What they need:
    - Real-time data synchronization across devices
    - Offline-first capabilities with automatic sync
    - Low latency for responsive user experience
    - Easy integration with mobile SDKs
- Best database choices:
    - Firebase Realtime Database – Built-in sync, offline support, real-time listeners
    - Firestore – More structured than Realtime Database, better querying


## c. AI-Powered Assistants (Chatbots, RAG Systems)
- What they store:
    - Vector embeddings of text and documents
    - Document chunks for retrieval
    - Conversation history and context
    - Semantic relationships between content
- What they need:
    - Efficient vector similarity search (nearest neighbor)
    - Low-latency retrieval for real-time responses
    - Support for high-dimensional vectors
    - Integration with embedding models
- Best database choices:
    - ChromaDB – Purpose-built for embeddings, developer-friendly, open source
    - FAISS – Facebook's high-performance similarity search library
    - Pinecone or Weaviate – Managed vector database solutions

## d. Cybersecurity Monitoring Tools
- What they store:
    - Security logs and audit trails
    - Threat detection events
    - Attacker behavior patterns
    - Real-time alert data
- What they need:
    - High write throughput to handle massive log volumes
    - Flexible schema to accommodate evolving threat data
    - Fast querying across unstructured data
    - Horizontal scalability for growing data
- Best database choices:
    - MongoDB – Flexible document model, excellent write performance. Good for log heavy systems
    - Elasticsearch – Powerful search capabilities, built for log analysis


> **When unsure, ask following questions and your answers will guide your DB choice:**
> 1. How structured is my data?
> 2. Will my schema change often?
> 3. Do I need fast search or exact accuracy?
> 4. Am I building something real-time?

# <span style='background :lightgreen' >3. Connecting to Databases: Where They Live and How to Access Them</span>
## a. Deployment Models
### (i) Local Databases (On Your Machine)
- Location: Installed and running on your computer
- Connection: Direct connection via localhost or 127.0.0.1
- Examples: MySQL (running locally on port 3306), MongoDB (running locally on port 27017), SQLite (file on your filesystem)
- Pros: Fast, no internet needed, full control
- Cons: Not accessible remotely, limited resources
### (ii) Cloud-Hosted Databases (Database-as-a-Service)
- Location: Managed by cloud providers
- Connection: Remote connection via internet using URLs/endpoints
- Examples: MongoDB Atlas, Firebase/Firestore (Google Cloud), AWS RDS, Azure SQL Database, Elastic Cloud
- Pros: Scalable, managed backups, high availability
- Cons: Requires internet, ongoing costs, potential latency
### (iii) Self-Hosted Remote Databases
- Location: Your own server/VPS
- Connection: Remote connection via IP address or domain
- Examples: MongoDB on AWS EC2, PostgreSQL on DigitalOcean
- Pros: Full control, customizable
- Cons: You manage maintenance, security, backups

### Essential Connection Parameters
- Host/Endpoint: Where the database is located
- Local: localhost or 127.0.0.1
- Remote: IP address or domain name (e.g., cluster0.mongodb.net)
- Port: Network port the database listens on
    - MySQL: 3306
    - PostgreSQL: 5432
    - MongoDB: 27017
    - Elasticsearch: 9200
- Credentials:
    - Username: Database user account
    - Password: Authentication password
    - API Keys: For cloud services (Firebase, managed services)
- Database Name: Specific database to connect to (if applicable)
- Additional Parameters (varies by database):
    - SSL/TLS settings for secure connections
    - Connection timeouts
    - Connection pool size
    - Authentication database (MongoDB)

#### DDL (Data Definition Language)
- Commands for creating, modifying, dropping and renaming the structure of database objects.
- **DDL = `CREATE`, `ALTER`, `DROP`, `TRUNCATE`, `RENAME`**
    - `CREATE` - Create database, table, index, view
    - `ALTER` - Modify table structure (add/modify/drop columns)
    - `DROP` - Delete database, table, index, view
    - `TRUNCATE` - Remove all records from table (keeps structure)
    - `RENAME` - Rename table or column
```sql
-- Create a table
mysql> CREATE TABLE users (
    id INT AUTO_INCREMENT PRIMARY KEY,
    first_name VARCHAR(50) NOT NULL,
    last_name VARCHAR(50) NOT NULL,
    email VARCHAR(100) UNIQUE,
    created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);
-- View all tables in current database
mysql> SHOW TABLES;
-- View table structure
mysql> DESCRIBE users;
-- or
mysql> SHOW COLUMNS FROM users;
-- Add a column to existing table
mysql> ALTER TABLE users ADD COLUMN address VARCHAR(100);
-- Modify a column
mysql> ALTER TABLE users MODIFY COLUMN address VARCHAR(200);
-- Rename a column
mysql> ALTER TABLE users RENAME COLUMN address TO street_address;
-- Drop a column
mysql> ALTER TABLE users DROP COLUMN address;
-- Rename a table
mysql> RENAME TABLE users TO members;
-- or
mysql> ALTER TABLE users RENAME TO members;
-- Delete all data from table (keeps structure)
mysql> TRUNCATE TABLE users;
-- Delete table completely
mysql> DROP TABLE users;
-- Create table with foreign key
mysql> CREATE TABLE orders (
    order_id INT AUTO_INCREMENT PRIMARY KEY,
    user_id INT,
    order_date DATE,
    FOREIGN KEY (user_id) REFERENCES users(id)
);
```
#### DML (Data Manipulation Language)
- Commands for retrieving, inserting, updating and deleting data within tables.
- **DML = `SELECT`, `INSERT`, `UPDATE`, `DELETE`**
    - `SELECT` - Retrieve/query data from tables
    - `INSERT` - Add new records to a table
    - `UPDATE` - Modify existing records in a table
    - `DELETE` - Remove records from a table
```sql
-- Select all records
mysql>  SELECT * FROM users;
-- Select specific columns
mysql> SELECT first_name, last_name FROM users;
-- Select with WHERE clause
mysql> SELECT * FROM users WHERE first_name = 'Arif';
-- Select with multiple conditions
mysql> SELECT * FROM users WHERE first_name = 'Arif' AND last_name = 'Butt';
-- Select with OR condition
mysql> SELECT * FROM users WHERE first_name = 'Arif' OR first_name = 'Rauf';
-- Select with LIKE (pattern matching)
mysql> SELECT * FROM users WHERE email LIKE '%@example.com';
-- DISTINCT values
SELECT DISTINCT last_name FROM users;
-- Select with ORDER BY
mysql> SELECT * FROM users ORDER BY last_name ASC;
-- Select with LIMIT
mysql> SELECT * FROM users LIMIT 5;
-- Select with OFFSET (skip first 5 records)
mysql> SELECT * FROM users LIMIT 5 OFFSET 5;
-- Count records
mysql> SELECT COUNT(*) FROM users;
-- JOIN tables
SELECT users.first_name, users.last_name, orders.order_date FROM users INNER JOIN orders ON users.id = orders.user_id;
-- GROUP BY with aggregate functions
mysql> SELECT last_name, COUNT(*) as count FROM users GROUP BY last_name;
-- HAVING clause (filter after GROUP BY)
mysql> SELECT last_name, COUNT(*) as count FROM users GROUP BY last_name HAVING count > 1;
-- Subquery
mysql> SELECT * FROM users WHERE id IN (SELECT user_id FROM orders);
-- Insert single record
mysql> INSERT INTO users (first_name, last_name, email) VALUES ('Arif', 'Butt', 'arif@example.com');
-- Insert multiple records
mysql> INSERT INTO users (first_name, last_name, email) VALUES ('Rauf', 'Khan', 'rauf@example.com'), ('Ahmed', 'Ali', 'ahmed@example.com');
-- Update single column value in a record
mysql> UPDATE users SET last_name = 'Khan' WHERE first_name = 'Arif';
-- Update multiple columns in a record
mysql> UPDATE users SET last_name = 'Khan', email = 'arif.khan@example.com' WHERE first_name = 'Arif';
-- Delete specific records
mysql> DELETE FROM users WHERE first_name = 'Arif' AND last_name = 'Butt';
-- Delete all records (use with caution!)
mysql> DELETE FROM users;
```
#### DCL (Data Control Language)
- Commands for controlling access and managing permissions on database objects.
- **DCL = `GRANT`, `REVOKE`**
    - `GRANT` - Give specific privileges to users
    - `REVOKE` - Remove privileges from users
```sql
-- Create a new user
mysql> CREATE USER 'arif'@'localhost' IDENTIFIED BY 'password123';
```

#### TCL (Transaction Control Language)
- Commands for managing transactions and maintaining data consistency and integrity.
- **TCL = `START TRANSACTION` (`BEGIN`), `COMMIT`, `ROLLBACK`, `SAVEPOINT`**
    - `START TRANSACTION` - Start a new transaction
    - `COMMIT` - Save all changes made in the current transaction
    - `ROLLBACK` - Undo all changes made in the current transaction
    - `SAVEPOINT` - Create a point within a transaction to rollback to
- Within YOUR session, you can see uncommitted changes immediately. This is called "dirty reads" within your own transaction. However, from OTHER sessions/connections, users cannot see your changes until you COMMIT. They see the data as it was before your transaction started 
```sql
-- Example transaction that ends with a commit 
mysql> START TRANSACTION;
    Imysql> NSERT INTO users (first_name, last_name) VALUES ('Test', 'User');
    mysql> UPDATE users SET last_name = 'Updated' WHERE first_name = 'Test';
    mysql> DELETE FROM users WHERE first_name = 'Old';
mysql> COMMIT;  -- Save all changes

-- Example transaction that ends with a Rollback (undo changes)
mysql> START TRANSACTION;
    mysql> DELETE FROM users WHERE id = 1;
mysql> ROLLBACK;  -- Undo the delete or all operations after the start of the transaction

-- Using SAVEPOINT
START TRANSACTION;
    INSERT INTO users (first_name, last_name) VALUES ('User1', 'Test');
    SAVEPOINT sp1;
    INSERT INTO users (first_name, last_name) VALUES ('User2', 'Test');
    SAVEPOINT sp2;
    INSERT INTO users (first_name, last_name) VALUES ('User3', 'Test');
    -- Rollback to sp1 (keeps User1, removes User2 and User3)
    ROLLBACK TO sp1;
COMMIT;
```

# <span style='background :lightgreen' >5. A  Quick Comparison of MySQL and SQLite Databases</span>
- **MySQL** is a full-featured client-server database system designed for multi-user applications, while **SQLite** is a lightweight, embedded database that runs as part of your application with zero configuration.
- Analogy:
    - MySQL = Restaurant with waiters (clients connect to a central server)
    - SQLite = Packed lunch (database travels with your app)

| Feature / Aspect | MySQL | SQLite |
|------------------|-------|--------|
| **Architecture** | Client-Server (separate server process) | Embedded/Serverless (runs within application) |
| **Installation** | Requires server installation & configuration | No installation - just a library (~600 KB) |
| **Setup Complexity** | Needs configuration, user setup, server start | Zero configuration - just specify file path |
| **Storage Model** | Multiple files in data directory | Single `.db` or `.sqlite` file |
| **Connection** | Requires host, port, username, password | Just file path - `sqlite3.connect('mydb.db')` |
| **Data Types** | Many specific types (INT, VARCHAR, DATE, etc.) | Flexible - only 5 storage classes |
| **Portability** | Database tied to server location | Database file can be copied/moved anywhere |
| **Concurrent Users** | Thousands of simultaneous connections | Limited - one write at a time, multiple reads |
| **Write Concurrency** | Multiple simultaneous writes | Only one write operation at a time |
| **User Management** | Built-in authentication & authorization | No user management - file permissions only |
| **Security** | Username/password, RBAC, network security | Operating system file permissions |
| **Performance** | Optimized for network & multi-user access | Optimized for local disk access & speed |
| **Best For** | Web apps, multi-user systems, production | Mobile apps, desktop apps, prototypes, embedded systems |
| **Common Use Cases** | E-commerce sites, SaaS platforms, CMS (WordPress) | Mobile apps (WhatsApp), browsers (Chrome), IoT devices |
| **Python Library** | `mysql-connector-python`, `pymysql` | `sqlite3` (built into Python) |
| **Deployment** | Requires server infrastructure | Deploys with application |

## Create a new database with the name of `student_db1`
- In SQLite, each database is a separate file. To create a new database, you simply connect to a new file.
- The `sqlite3.connect()` method creates and returns a connection with a database object.
    - If the file/db donot exist SQLite automatically creates the database file.
    - The connection object handles all the communication between the Python program and the database
    - Think of it as the "phone line" to the database
    - No host, user, or password needed (SQLite is file-based),
- The `con.cursor()` method creates and returns a cursor object that is used to work with that connection:
    - A control structure that lets you navigate through and manipulate query results
    - Used to execute SQL commands and fetch results
    - Think of it as a "pointer" or "worker" that does the actual work on that connection
    - You can have multiple cursors on a single connection
    - Think of it as an "employee" on the other end of "phone line", who takes your requests and gives you responses

````python
import sqlite3
import os

# Delete the database file if it exists (No need for DROP DATABASE as in case of MySQL)
if os.path.exists("student_db1.db"):
    os.remove("student_db1.db")
    print("Old database deleted.")

# Create a new database
conn = sqlite3.connect("student_db1.db")

# Create a cursor object that is used to execute SQL commands and fetch results
cursor = conn.cursor()
print("Database 'student_db1.db' created successfully!\n")
````

## Create a Table students in the database student_db
- In SQLite, you're already working with the specific database file you connected to
- There's no need to switch databases like in MySQL since each connection works with one database file
- SQLite uses INTEGER PRIMARY KEY for auto-increment (it's automatic, no AUTOINCREMENT keyword needed)

````python
# Execute SQL command to create 'students' table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS students (
        id INTEGER PRIMARY KEY,  -- Auto-incrementing primary key (INTEGER in SQLite)
        name TEXT,               -- Student name (TEXT for strings in SQLite)
        age INTEGER              -- Student age (INTEGER data type)
    )
''')

print("Table 'students' created successfully!\n")

# Commit the table creation to save it permanently
conn.commit()
````
