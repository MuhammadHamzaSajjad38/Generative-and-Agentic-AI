## a. Overview of Embeddings
<h3 align="center"><div class="alert alert-success" color=magenta style="margin: 20px">Word Embedding is a technique, where each word/token is represented as a dense n-dimensional numerical vector that encode the semantic and contextual meaning of the token. Words having similar meanings are positioned closer together, enabiling models to perform tasks like similarity measurement, semantic search, clustering and information retrieval. </h3>


- The core idea is to represent words as **dense, low-dimensional vectors** that capture **semantic meaning** and **distributional similarity** to overcome the limitations of sparse, high-dimensional representations used in statistical and early ML methods.
- **Embeddings** are high-dimensional vectors that capture the semantic meaning of words, sentences or even entire documents. The process of converting text into embeddings allows LLMs to perform various natural language processing tasks, such as text generation, sentiment analysis and more.
- Operations on vectors, such as a dot product, help us discover whether two vectors are identical or different. At a high level, this forms the basis for performing similarity search on vectors stored in memory or in specialized vector databases.
- Through embeddings, modern NLP models achieve a deep understanding of language, enabling tasks like sentiment analysis, text summarization, and question answering with nuanced comprehension and generation capabilities. They serve as the foundation for LLMs, but are also used independently to convert text into vectors while retaining semantic context. When text is passed through an embedding model, a vector is produced that contains the embeddings.

## b. Static Embedding Models
<h3 align="center"><div class="alert alert-success" color=magenta style="margin: 20px">Static embeddings generate word vectors that assign the same fixed representation to a word every time it appears, regardless of context.</h3>

- **Word2Vec (Google, 2013, https://arxiv.org/pdf/1301.3781.pdf):** Creates static word embeddings by predicting neighboring words in a fixed context window, producing the same vector for each word regardless of context. It learns embeddings using neural networks with two main architectures:
    - *Continuous Bag of Words (CBOW)* Take into consideration the words that are around a given target words and the goal is the predict the target words. (Predicts a word from its surrounding context)
    - *Skip-Gram* is the opposite, in which you predicts surrounding context words from a given word.
- **GloVe (Stanford, 2014, https://nlp.stanford.edu/pubs/glove.pdf):** Combines global matrix factorization with local context window methods to create word vectors that capture both global statistical information and local context patterns. It leverages word co-occurrence statistics from the entire corpus to learn embeddings that perform well on word analogy tasks. GloVe stands for "Global Vectors for Word Representation" and often achieves competitive performance with Word2Vec while being more computationally efficient to train.
- **FastText (Facebook, 2016, https://arxiv.org/pdf/1607.01759.pdf):** Extends Word2Vec by incorporating subword information (character n-grams), making it better at handling out-of-vocabulary words and morphologically rich languages. It represents each word as a bag of character n-grams, allowing it to generate embeddings for words not seen during training by combining subword representations. This approach is particularly effective for languages with rich morphology like Turkish, Finnish, or German where word forms can vary significantly.

### Limitations of static embeddings (NO Context Awareness):
- The same word always gets the same vector
- In sentences, **"I deposited a cheque in bank"** and **"I slipped at the river bank"**, the word bank will get the same embeddings, making it impossible to distinguish between the two meanings.
- In sentences, **"This apple is red and juicy"** and **"Apple launched a new phone"** the word apple will get the same embeddings. In the first sentence the word apple refer to a fruit, while in the second sentence the word apple refers to a company.

>- If the training data on which the embedding model is trained has 1000 occurrances of word apple as fruit and 20 occurrances of word apple as company. The resulting static embedding will be inclined  towards treating word "apple" as fruit.

## f. Contextual Embedding Models and Language Models Paradigms
<h3 align="center"><div class="alert alert-success" color=magenta style="margin: 20px">Contextual embeddings generate word vectors that dynamically change based on the surrounding context, capturing different meanings for the same word in different sentences.</h3>
    
- Modern Natural Language Processing (NLP) systems rely on contextual embeddings, where the meaning of a word is not fixed but instead adapts based on its surrounding context.
- Early breakthroughs in contextual representation were achieved using LSTM-based architectures, most notably **ELMo**, which introduced bidirectional context modeling over sequences.
- The field rapidly evolved with the introduction of transformer architectures, which use self-attention to capture long-range dependencies and process entire sequences in parallel.
- Transformers enabled far richer representations and scalable training, leading to the diverse ecosystem of language models used today.
- Based on what they are optimized to do, modern language models can be broadly organized into three overlapping paradigms:
    - **Representational models** focus on learning rich contextualized representations (embeddings) of text that capture meaning, syntax, and relationships between words, without primarily aiming to generate new text
    - **Generative models** focus on modeling the probability of text sequences and generate new text token by token, usually in an autoregressive (left-to-right) manner, enabling them to produce coherent paragraphs, dialogues, code, or summaries.
    - **Hybrid models** explicitly combine both representation learning and text generation within a single architecture.

# <span style='background :lightgreen' >2. Overview of Vector Databases</span>
<h3 align="center"><div class="alert alert-success" color=magenta style="margin: 20px">A vector database is a specialized system that persistently stores and efficiently searches high-dimensional vectors to find semantically similar data at scale with enterprise features.</h3>

## a. Key Features of a Vector Database
- **Storage:** Ensures that vectors and their associated metadata are retained, whether **In-memory** for quick lookups or **On-disk** for durability and large-scale use. Vector databases don't just hold the numerical vectors - they also maintain the original data (text, images, etc.) alongside the vector representations, creating a complete searchable database.
- **Similarity Search:** Helps retrieve the vectors most similar to a query vector. Instead of exact matches, vector stores use mathematical distance calculations (like cosine similarity or Euclidean distance) to find the "closest" vectors in the high-dimensional space. This enables finding conceptually similar content even when the exact words or features differ.
- **Indexing:** Provides a data structure or method that enables fast similarity searches on high-dimensional vectors (e.g., approximate nearest neighbor lookups). With millions of vectors, searching through each one individually would be too slow. Indexing creates optimized structures like Hierarchical Navigable Small World (HNSW) graphs, Inverted File (IVF) indexes and Locality Sensitive Hashing (LSH), which allow sub-second searches even across massive datasets.
- **CRUD Operations:** Manages the lifecycle of data - adding new vectors, reading them, updating existing entries, removing outdated vectors. Just like traditional databases, vector stores need to handle:
    - **Create**: Add new vector embeddings
    - **Read**: Retrieve vectors and perform searches
    - **Update**: Modify existing vector data
    - **Delete**: Remove obsolete or unwanted vectors

## b. Common Use Cases of Vector Databases
- **Semantic Search Engines:** Enables searching by meaning rather than keywords — ideal for finding conceptually related content. For example, a researcher searching for “eco-friendly transport” also finds articles about “electric vehicles” or “sustainable mobility.”
- **Chatbots & QnA Systems:** Retrieves the most relevant information from a knowledge base to answer user questions accurately. For example, a customer support bot fetching exact policy details or troubleshooting steps from documentation.
- **RAG (Retrieval-Augmented Generation):** Enhances AI responses by supplying relevant context retrieved from a vector store before generation. For example, An AI assistant that accesses company manuals or medical literature to give factual, up-to-date answers.
- **Recommendation Systems:** Suggests items with similar semantics or user preference patterns instead of just matching categories. For example, Spotify recommending songs with a similar mood or energy level, even if they belong to different genres. Similarly, Netflix analyze your viewing history to recommend movies with similar themes, moods, or storytelling styles - not just the same genre. Similarly, various E-commerce sites use this "Customers who bought this also liked...". Similarly, different Social media platoforms show only relevant posts and content in your feed.
- **Image/Multimedia Search:** Finding similar images, audio, or video content based on visual or audio features. For example upload a photo and find visually similar images, or search for music with similar rhythm and mood patterns.

## c. How Do Vector Databases Work?
- **Chunking the Data:** Raw content (text, images, audio, etc.) is divided into smaller, meaningful segments called chunks for more precise semantic retrieval.
- **Generating Embeddings:** Each chunk is converted into a high-dimensional vector (a list of numbers) using an AI embedding model. These vectors represent the meaning or context of the data rather than exact words or pixels.
- **Storage & Indexing:** The vectors are stored inside the vector database and organized using specialized indexing structures (e.g., HNSW, IVF, FAISS) to enable fast similarity searches.
- **Query Embedding:** When a user submits a query, it is also transformed into an embedding vector using the same model.
- **Similarity Search (ANN):** The system finds the closest vectors to the query using Approximate Nearest Neighbor (ANN) algorithms, typically based on cosine similarity, dot product, or Euclidean distance.
- **Result Retrieval:** The database returns the original data (text passages, images, etc.) linked to the most similar vectors — providing semantically relevant results instead of keyword matches.

## d. Vector Store vs Vector Database
- **Vector Store:** A vector store is a lightweight library or in-memory index that temporarily holds vectors and performs similarity searches using algorithms like FAISS or Annoy. It's ideal for prototyping and experimentation but lacks persistence, meaning data is lost when the program stops, and it doesn't support advanced features like filtering, updates, or multi-user access.
- **Vector Database:** A vector database is a production-grade system designed to persistently store vectors and efficiently search through millions or billions of them using specialized indexing techniques. Unlike traditional databases that match exact keywords, vector databases find data that is similar in meaning through semantic search, while also providing enterprise features like CRUD operations, metadata filtering, scalability, security, and high availability.
| Aspect | Vector Store | Vector Database |
|--------|--------------|--------------------------------------|
| **Definition** | Lightweight library or in-memory index for temporary vector storage | Specialized system that stores embeddings persistently and enables nearest-neighbor search |
| **Storage** | In-memory only, RAM-based | Multiple options: in-memory, local disk, cloud, or databases like PostgreSQL |
| **Data Persistence** | ❌ Lost when program ends | ✅ Data survives program restarts |
| **Search Capability** | Similarity search using FAISS/Annoy algorithms | Vector search + full-text search + metadata filtering + multi-modal retrieval |
| **Scale** | Single machine (millions of vectors max) | Scalable: development (100K entries) to production (billions of vectors) |
| **Use Cases** | Prototyping, testing, experimentation, learning | Development, production, enterprise applications |
| **CRUD Operations** | ❌ Limited or none | ✅ Full Create, Read, Update, Delete support |
| **Metadata Filtering** | ❌ No | ✅ Yes, advanced filtering capabilities |
| **Multi-User Access** | ❌ No (single process only) | ✅ Yes (client-server architecture) |
| **Production Ready** | ❌ No | ✅ Yes |
| **Deployment** | Embedded in Python scripts | Laptop, local machine, public/private cloud, Kubernetes |

## e. Types of Similarity Search in Vector Databases
- Vector DBs use different algorithms to find "similar" data based on embeddings. Below is a unified view of <b>Exact</b>, <b>ANN</b>, and <b>Hybrid</b> similarity techniques.
- Indexing is a technique used in vector databases to make similarity search faster and more efficient. Instead of comparing your query with every single vector (which is slow), indexing helps the system narrow down the search space using smart structures like graphs, clusters, or compressed codes.
- In **ChromaDB, indexing happens automatically** whenever you add documents. It uses a flat (brute-force) search index by default, meaning it compares each query embedding with all stored embeddings for maximum accuracy. While this is efficient for small to medium datasets, other vector databases like Qdrant, Milvus, and Weaviate support advanced Approximate Nearest Neighbor (ANN) indexes such as HNSW, IVF, and PQ for faster large-scale search.

  <td>Stores all vectors in memory and compares the query to every vector using exhaustive search; guarantees 100% accuracy with no approximations but scales linearly with dataset size.</td>
  <td>ChromaDB (default), FAISS IndexFlat, pgvector — ideal for small to medium datasets (up to ~100K vectors)</td>
</tr>

<!-- ANN Techniques -->
<tr style="background-color:#fef9e7;">
  <td><strong>HNSW (Hierarchical Navigable Small World)</strong></td>
  <td style="text-align:center; vertical-align:middle;">ANN / Index</td>
  <td>Builds a multi-layer graph structure where each layer contains progressively fewer nodes; search starts at the top layer and navigates down through nearest neighbors, achieving very fast approximate search with high recall (95%+ typical).</td>
  <td>Weaviate, Qdrant, Pinecone, Milvus, pgvector, Chroma (via hnswlib)</td>
</tr>
<tr style="background-color:#fef9e7;">
  <td><strong>IVF (Inverted File Index)</strong></td>
  <td style="text-align:center; vertical-align:middle;">ANN / Index</td>
  <td>Clusters vectors into partitions (Voronoi cells) using k-means; during search, only vectors in the nearest clusters are examined, significantly reducing search space and improving speed.</td>
  <td>FAISS (IndexIVFFlat), Milvus, ScaNN, AnalyticDB</td>
</tr>
<tr style="background-color:#fef9e7;">
  <td><strong>PQ (Product Quantization)</strong></td>
  <td style="text-align:center; vertical-align:middle;">ANN / Index</td>
  <td>Compresses high-dimensional vectors by dividing them into subvectors and quantizing each independently into small codes; dramatically reduces memory footprint (8-64x compression) while enabling fast approximate distance computations.</td>
  <td>FAISS (IndexPQ, IndexIVFPQ), Milvus, ScaNN</td>
</tr>
<tr style="background-color:#fef9e7;">
  <td><strong>Annoy (Approximate Nearest Neighbors Oh Yeah)</strong></td>
  <td style="text-align:center; vertical-align:middle;">ANN / Index</td>
  <td>Builds multiple random projection trees (binary trees) where each node splits vectors by a random hyperplane; queries traverse multiple trees in parallel for fast lookups. Optimized for static, read-heavy workloads.</td>
  <td>Spotify Annoy library, Milvus</td>
</tr>
<tr style="background-color:#fef9e7;">
  <td><strong>LSH (Locality-Sensitive Hashing)</strong></td>
  <td style="text-align:center; vertical-align:middle;">ANN / Index</td>
  <td>Uses hash functions designed so that similar vectors are likely to hash to the same bucket; enables sublinear search time by checking only vectors in matching buckets rather than the entire dataset.</td>
  <td>Research implementations, Redis with RediSearch, custom ANN engines</td>
</tr>
<tr style="background-color:#fef9e7;">

### STEP 3 (Optional): Create Embedding Function
- When you add a document using the `collection.add()` method, ChromaDB automatically generates embeddings using its default model `all-MiniLM-L6-v2` from Sentence Transformers.
- Python provides several embedding function wrappers in `chromadb.utils.embedding_functions.` like:
    - **`SentenceTransformerEmbeddingFunction()`:** SentenceTransformers is a Python library that provides access to local Hugging Face Sentence Transformers models (e.g., `all-MiniLM-L6-v2`, `all-mpnet-base-v2`, `paraphrase-multilingual-MiniLM-L12-v2`) specifically designed to generate semantic embeddings for sentences, paragraphs, or individual words. It encodes text into dense vector representations that can be used for similarity search, clustering, or other NLP tasks. You can download and run open-source embedding models from Hugging Face directly on your local machine (CPU/GPU). These models are free to use, with no API costs or rate limits.
    - **`OpenAIEmbeddingFunction()`:**. Uses OpenAI Embeddings (e.g., `text-embedding-3-small`, `text-embedding-3-large`) that are trained on large-scale, high-quality corpora and are optimized for semantic search, retrieval, clustering, classification, and recommendation systems. No model files are ever downloaded to your machineare generated via remote API calls to OpenAI’s managed servers. You simply send text to the API and receive back dense embedding vectors. The OpenAI embedding models are hosted, paid, large embeddings and are good for enterprise-level applications.
    - **`OllamaEmbeddingFunction()`:**. Uses models served by a local or remote Ollama server (e.g., `nomic-embed-text`). Great for self-hosted or privacy-focused setups.

````python
######################        DefaultEmbeddingFunction()       #############################

from chromadb.utils import embedding_functions

ef_miniLM = embedding_functions.DefaultEmbeddingFunction() 
text = ["Learning is fun with Arif"] # Text to embed
embeddings = ef_miniLM(text)       # Generate embeddings

print("First 5 embedding values: ", embeddings[0][:5])
print("\nEmbedding dimension:", len(embeddings[0]))
````

````python
######################        SentenceTransformerEmbeddingFunction()       #############################

from chromadb.utils import embedding_functions

ef_mpnet = embedding_functions.SentenceTransformerEmbeddingFunction(model_name="all-mpnet-base-v2") 
text = ["Learning is fun with Arif"] # Text to embed
embeddings = ef_mpnet(text)       # Generate embeddings

print("First 5 embedding values: ", embeddings[0][:5])
print("\nEmbedding dimension:", len(embeddings[0]))
````

## b. Read Operations

### Semantic Search using the `collection.query()` Method
- The `collection.query()` method is used to retrieve documents that are semantically similar to a query text or embedding.
- Use Case: When you want to find conceptually related documents (e.g., “What is FastAPI?” → finds text about Python web frameworks).
- Key Parameters:
    - `query_texts`: List of query strings (converted into embeddings).
    - `where`: (Optional) Filter by metadata before doing semantic search.
    - `n_results`: Number of top results to return.
    - `include`: Choose what to return — e.g., "documents", "metadatas", "distances".
- Returns: Ranked results with semantic distances (smaller = more similar).

>-  In ChromaDB, the `collection.query()` method is used to query one collection. There is no direct API to query all collections together, however, you can programmatically loop through each collection to do this

````python
# Performs semantic search over all documents in collection1. Finds the 3 most similar documents to the query "What is FastAPI?" using the default embedding function.
# Returns a dictionary with keys like 'documents', 'embeddings', and 'distances'.
results = collection1.query(
                            query_texts=["What is FastAPI?"],  # List of query strings (converted into embeddings).
                            n_results=3                        # Number of top results to return.
                            )

print("Results:")
# enumerate(): Adds a counter to each item, its second parameter 1 means starts counting from 1 instead of 0. It returns a Tuples of (index, item)
# zip(): Combines two lists element by element and creates: Pairs of (document, distance)
for i, (doc, distance) in enumerate(zip(results['documents'][0], results['distances'][0]), 1):
    print(f"{i}. {doc}")
    print(f"   Distance: {distance:.4f}\n")
````

````python
# Searches for the 3 most relevant documents that match "AI". Additionally filters results using where={"category": "programming"}. Only documents in collection1 where category equals "programming" are considered.
# Useful for metadata-based filtering.
results = collection1.query(
                            query_texts=["AI"],                # List of query strings (converted into embeddings).
                            where={"category": "programming"}, # Filter by metadata before doing semantic search.
                            n_results=3                        # Number of top results to return.
                            )

print("Results:")
for i, (doc, distance) in enumerate(zip(results['documents'][0], results['distances'][0]), 1):
    print(f"{i}. {doc}")
    print(f"   Distance: {distance:.4f}\n")
````

````python
# Only considers documents in collection1 with category = programming AND framework = FastAPI. Useful when metadata is hierarchical.
results = collection2.query(
                            query_texts=["Greatest football player"], # List of query strings (converted into embeddings).
                            n_results=2,                              # Number of top results to return.
                            include=["documents", "distances"]        # Limit returned fields using include (documents and distances only)
                            )

print("Results:")
for i, (doc, distance) in enumerate(zip(results['documents'][0], results['distances'][0]), 1):
    print(f"{i}. {doc}")
    print(f"   Distance: {distance:.4f}\n")
````

### Get Statistics of Collections
- **`client.list_collections()`:**
    - The `client.list_collections()` method is used to return a list of all collections currently stored in your Chroma database.
    - Each collection object typically includes attributes such as name, id, and optional metadata.
    - Returns a list of collection objects.
    - Use Case: When you need to view or iterate through all existing collections in the database for inspection or management.

### Direct Retrieval using the `collection.get()` Method 
- The `collection.get()` method is used to retrieve documents exactly matching certain filters (no semantic similarity).
- Use Case: When you know the document IDs or want to fetch data by metadata conditions.
- Key Parameters:
    - `ids`: List of specific document IDs to fetch.
    - `where`: Metadata filter (e.g., {"category": "sports"}).
    - `limit`: Max number of results to return.
- Returns: Exact documents, embeddings, and metadata that match — without ranking or distance scores.

````python
def show_collection_docs(collection):
    print(f"\n📂 Documents in Collection: '{collection.name}'")
    all_docs = collection.get()
    for i, (doc_id, doc, metadata) in enumerate(
        zip(all_docs['ids'], all_docs['documents'], all_docs['metadatas']), 1
    ):
        print(f"   {i}. [{doc_id}] {doc}")
        print(f"      Metadata: {metadata}")

show_collection_docs(collection1)
show_collection_docs(collection2)
````

## c. Update Operations
- The `collection.update()` method is used to modify existing documents, embeddings, or metadata within a collection.
- Use Case: When you want to correct, refine, or enrich previously added documents without re-creating the entire collection.
- Key Parameters:
    - `ids`: List of document IDs to identify which records to update.
    - `documents`: New text content to replace the old one.
    - `metadatas`: Updated metadata fields (you can add new keys or modify existing ones).
- Returns: Confirms that the specified records have been updated in the collection.

````python
collection1.update(
                    ids=["c1doc1"],
                    documents=["Python is a versatile programming language widely used in AI, web development, and data science."],
                    metadatas=[{"category": "programming", "language": "Python", "updated": True}]
                    )

# Verify update
updated = collection1.get(ids=["c1doc1"])
print(f"Updated document: {updated['documents'][0][:100]}...")
print(f"   Metadata: {updated['metadatas'][0]}")
````

## d. Delete Operations
- The `collection.delete()` method is used to remove documents, embeddings, or metadata entries from a collection.
- Use Case: When you want to delete specific documents, clean up outdated data, or remove items matching certain metadata conditions.
- Key Parameters:
    - `ids`: List of document IDs to delete, used to remove a specific document(s).
    - `where`: Metadata filter (e.g., {"category": "sports"}) to delete all matching records.
- Returns: Confirms successful deletion of the specified documents and their associated embeddings from the collection.

````python
# Delete a specific document by ID
collection1.delete(ids=["c1doc4"])
print("   • Deleted document with ID = c1doc4")

# Delete documents by metadata filter
collection2.delete(where={"sport": "Cricket"})
print("   • Deleted all documents where sport = 'Cricket'")

# VERIFY DELETION

remaining_c1 = collection1.get()
remaining_c2 = collection2.get()

print("\n📊 Remaining Documents:")
print(f"   • Collection 1 ({collection1.name}): {len(remaining_c1['documents'])} documents")
print(f"     IDs: {remaining_c1['ids']}")
print(f"   • Collection 2 ({collection2.name}): {len(remaining_c2['documents'])} documents")
print(f"     IDs: {remaining_c2['ids']}")
````

# <span style='background :lightgreen' >6. Understanding Storage Structure of ChromaDB</span>
## a. Understand what all files are saved and where?
```
chroma_db2/
 ├── chroma.sqlite3            →  relational database used by Chroma to store collection definitions, document IDs, metadata, text documents, mapping between records and embeddings, configuration
 │    ├── collections
 │    ├── embeddings
 │    ├── embedding_metadata
 │    └── segments
 │    └── segment_metadata
 │
 ├── 6b3c9c7e-.../             →  Each collection gets a UUID directory because Chroma builds a separate vector index for every collection
 │    ├── data_level0.bin      →   stores embedding vectors
 │    ├── header.bin           →   stores index metadata for HNSW structure
 │    ├── length.bin           →   stores vector length info and offset
 │    └── link_lists.bin       →   stores HNSW graph connections
 │
 └── a91fd2e1-.../      # another collection
      ├── data_level0.bin      →   stores embedding vectors
      ├── header.bin           →   stores index metadata for HNSW structure
      ├── length.bin           →   stores vector length info and offset
      └── link_lists.bin       →   stores HNSW graph connections
```
- This design separates:
    - SQLite database → metadata + documents
    - Binary index files → vector search index
