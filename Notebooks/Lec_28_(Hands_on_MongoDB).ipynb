## a. Comparison of SQL, NoSQL, and Vector Databases


| Feature | SQL Database (Relational) | NoSQL Database (Non-Relational) | Vector Database (AI-Native) |
|---------|---------------------------|----------------------------------|------------------------------|
| **Data Structure** | Structured tables with predefined schemas (rows and columns) | Flexible: documents, key-value pairs, graphs, or wide-column stores | High-dimensional vectors (embeddings) with optional metadata |
| **Data Type** | Text, numbers, dates, booleans stored in fixed columns | JSON/BSON documents, nested objects, arrays, varied data types | Numerical arrays (vectors) representing semantic meaning of data |
| **Schema** | Strict schema required before inserting data | Schema-less or dynamic schema (flexible structure) | Minimal schema: vectors + optional metadata (schema-flexible) |
| **Relationships** | Strong relationships using foreign keys and JOINs | Embedded documents or references (denormalization preferred) | Weak relationships; focuses on content similarity rather than explicit links |
| **Scalability** | Vertical scaling (bigger server); horizontal scaling is complex | Horizontal scaling (across  servers/shards); designed for distributed systems | Horizontal scaling; optimized for high-dimensional vector operations |
| **ACID Compliance** | Strong ACID guarantees (Atomicity, Consistency, Isolation, Durability) | Often eventual consistency; some offer ACID for single documents | Varies; typically prioritizes performance over strict ACID |
| **Query Language** | SQL (Structured Query Language) | Varies by database: MongoDB Query Language, CQL, Cypher, etc. | Similarity search APIs, distance metrics (cosine, euclidean, dot product) |
| **Query Style** | • Exact match queries<br>• Filtering, sorting, joining<br>• Example: `WHERE name = 'Ali'` | • Document/field queries<br>• Range queries, text search<br>• Example: `{name: "Ali", age: {$gt: 25}}` | • Semantic similarity search<br>• Nearest neighbor search<br>• Example: "Find documents similar to 'machine learning'" |
| **Query Example** | `SELECT * FROM users WHERE age > 25 AND city = 'Lahore'` | `db.users.find({age: {$gt: 25}, city: "Lahore"})` | `collection.query(query_texts=["young professionals in Lahore"])` |

### What is mongosh:
- `mongosh` (MongoDB Shell) is a command-line interface tool that allows you to interact with MongoDB databases through text commands
- It provides a JavaScript-based environment for running queries, managing databases, and performing administrative tasks
### Authentication in MongoDB:
- When you install MongoDB via Homebrew, it has **no default users** and authentication is **disabled by default**
- Therefore, you can connect without any username/password
### How to connect:
- Open a new terminal window
- Type the command: `mongosh`
- Press Enter to connect to MongoDB running on `localhost:27017`
- You'll see the prompt `test>` indicating you're connected to the default **test** database
- To exit mongosh, use the command: `exit` or `quit()`
### Default databases in MongoDB:
- **admin:** The authentication and authorization database where administrative users and system-wide commands are stored
- **config:** Stores metadata and configuration settings for sharded clusters (routing information, chunk distribution, etc.)
- **local:** Contains replication-specific data and is never replicated across replica set members; each server has its own local database
### What you can do in mongosh:
- **Create and switch databases:** Use `use databaseName` to create or switch to a database
- **Run queries:** Execute find, insert, update, and delete operations on collections
- **Manage collections:** Create, drop, and rename collections
- **User management:** Create users, assign roles, and manage permissions
- **Run JavaScript:** Execute JavaScript code and functions directly in the shell
- **Administrative tasks:** Monitor server status, manage indexes, and perform backups

#### Common Roles in MongoDB
```javascript
// Database User Roles:
- read              // Read data from all non-system collections
- readWrite         // Read and write data to all non-system collections
// Database Administration Roles:
- dbAdmin           // Perform administrative tasks
- userAdmin         // Create and modify roles and users
- dbOwner           // Perform any administrative action
// Cluster Administration Roles:
- clusterAdmin      // Manage and monitor the cluster
- clusterManager    // Manage and monitor operations
- clusterMonitor    // Read-only access to monitoring tools
// All-Database Roles:
- readAnyDatabase   // Read data from all databases
- readWriteAnyDatabase  // Read and write to all databases
- userAdminAnyDatabase  // Manage users on all databases
- dbAdminAnyDatabase    // Database admin on all databases
// Superuser Roles:
- root              // Complete access to all resources
```

## c. MongoDB Query Language (MQL)  - Insert Operations using `mongosh`
- MongoDB uses MQL (MongoDB Query Language) to perform insert operations through shell methods like `insertOne()` and `insertMany()`.
- These methods allow you to add new documents into a collection in a flexible, schema-less JSON format (internally stored as BSON).
```javascript
db.collection.insertOne(document)                      // Inserts a single document and returns an object containing the newly inserted document’s _id
db.collection.insertMany([document1, document2, ...])  // Inserts multiple documents and returns an object with an array of all inserted IDs
```
- The document is a JSON object enclosed in curly braces { }, representing the data to be inserted into the collection.
- Each inserted document automatically gets a unique _id field (ObjectId) if not provided.

## d. MongoDB Query Language (MQL) - Read Operations using `mongosh`
- MongoDB uses MQL (MongoDB Query Language), a JSON-like syntax for querying and manipulating data directly in the shell using methods such as `find()` and `findOne()`.
- The `find()` method returns a cursor that can be iterated or viewed using `.pretty()` for formatted output
- The `findOne()` method returns a single document that matches the filter (or null if none found)

```
db.collection.find()                     // No arguments → returns all documents
db.collection.find(filter)               // One argument (query/filter)
db.collection.find(filter, projection)   // Two arguments (filter + projection)
```
- The **filter** is a JSON-style object that specifies which documents to match, using query operators like `$gt`, `$lt`, `$in`, `$or`, etc. An empty {} means "match all documents"
- The **projection** specifies which fields to include (1) or exclude (0) from the results. By default, the `_id` field is included unless explicitly excluded using `"_id": 0`

### (i) BASIC `find()` and `find_one()` EXAMPLES
```bash
book_store_A> db.books.find()          # Example 1: Find all documents in books collection
dbook_store_A> db.customers.find()     # Example 2: Find all documents in customer collection
book_store_A> db.customers.findOne()   # Example 3: Find first document in customers collection
```

### (ii) PROJECTION (Selecting specific fields)
```bash
book_store_A> db.customers.find({}, {"name": 1, "age": 1})            # Example 4: Include only specific fields (1 = include)
book_store_A> db.customers.find({}, {"_id": 0, "name": 1, "age": 1})  # Example 5: Include only specific fields (1 = include and 0 = exclude)
book_store_A> db.books.find({}, {"_id": 0, "price": 0})               # Example 6: Exclude specific fields (0 = exclude)
```

### (iii) COMPARISON OPERATORS (`$gt`, `$gte`, `$lt`, `$lte`, `$ne`)
```bash
book_store_A> db.books.find({"author": "Dr. Arif"})                                  # Example 7: Find with simple filter (equality)
book_store_A> db.books.find({"author": {"$ne": "Dr. Arif"}})                         # Example 8: Not equal ($ne)
book_store_A> db.books.find({"author": "Dr. Arif"}, {"_id":0, "title":1, "price":1}) # Example 9: Find with simple filter (equality)
book_store_A> db.books.find({"price": {"$gte": 30}})                                 # Example 10: Greater than or equal ($gte)
book_store_A> db.books.find({"price": {"$gte": 25, "$lte": 35}})                     # Example 11: Range query (combine $gte and $lte)
book_store_A> db.customers.find({"age": {"$gt": 40}})                                # Example 12: Customers age > 40
```

### (iv) LOGICAL OPERATORS (`$and`, `$or`, `$not`, `$nor`)
```bash
book_store_A> db.books.find({"$and": [{"price": {"$gt": 25}}, {"author": "Dr. Jamil"}]})   # Example 13: $and operator (multiple conditions)
book_store_A> db.books.find({"$or": [{"author": "Dr. Arif"}, {"author": "Dr. Rauf"}]})     # Example 14: $or operator (at least one condition true)
book_store_A> db.books.find({"$or": [{"price": {"$lt": 30}}, {"price": {"$gt": 35}}]})     # Example 15: $or with price conditions
book_store_A> db.books.find({"price": {"$not": {"$gt": 30}}})                              # Example 16: $not operator
```

### (v) ARRAY OPERATORS (`$in`, `$nin`)
```bash
book_store_A> db.books.find({"author": {"$in": ["Dr. Arif", "Dr. Rauf"]}})    # Example 17: $in operator (match any value in array)
book_store_A> db.books.find({"price": {"$in": [25, 40]}})                     # Example 18: $in with numbers
book_store_A> db.books.find({"author": {"$nin": ["Dr. Arif", "Dr. Jamil"]}})  # Example 19: $nin operator (not in array)
book_store_A> db.orders.find({"books": "book_101"})                           # Example 20: Find orders containing specific book
```

## g. Indexing
- **What is Indexing?** Indexing in MongoDB improves the efficiency of search operations by allowing the database to locate documents without scanning every one. It's like an index in a textbook.
- **Benefits of Indexing:**
    - Faster queries
    - Improved performance for large datasets
    - Reduced resource usage
    - Indexing also helps in sorting
- **Trade-Offs of Indexing:**
    -  Indexes require additional memory and disk to store.
    -  Insert, update, and delete operations can become slower, as indexes need to be updated each time.
- **When not to use indexing in mongodb:**
    - When the collection is small
    - When the collection is frequently updated
    - When the queries are complex(multiple fields)
    - When the collection is large(make less indices)

- **Types of Indexes in MongoDB:**

| Index Type | Description | Advantage | When to Use | Example |
|------------|-------------|-----------|-------------|---------|
| **Single Field Index** | Creates an index on a single field of a document | Fast queries on one specific field | When filtering or sorting by a single field frequently | `db.users.createIndex({ email: 1 })` - Quick lookups by email |
| **Compound Index** | Creates an index on multiple fields in a specific order | Efficient queries that filter or sort by multiple fields together | When queries use multiple fields in WHERE clauses or sorting | `db.users.createIndex({ lastName: 1, firstName: 1 })` - Search by last name and first name |
| **Text Index** | Creates a full-text search index on string content | Enables searching for words/phrases within text fields | When implementing search functionality across text content | `db.articles.createIndex({ title: "text", content: "text" })` - Search articles by keywords |

### (i) Single Field Index
- A single-field index is created on one field of a collection and is useful when your queries frequently filter or sort by that field.
- Suppose we often search for books by their author. Without an index, MongoDB would scan every document (collection scan), while with the index on author, it can jump directly to matching entries
```bash
book_store_A> db.books.createIndex({"author": 1})         # creates an ascending index on the author field. (1 = ascending order)
book_store_A> db.books.find({"author": "Dr. Arif"})        # Such queries will now be faster
book_store_A> db.books.find({"author": "Dr. Arif"}).explain("executionStats")
```

### (ii) Compound Index
- A compound index includes multiple fields in a single index, and is useful when queries involve filtering or sorting on more than one field.
- Suppose we often search for books by both author and price. Without an index, MongoDB would scan every document (collection scan), while with the index on author and price, it can jump directly to matching entries
- Order matters in compound indexes — {author: 1, price: -1} is different from {price: -1, author: 1}. Design them based on the most common query patterns.
```bash
book_store_A> db.books.createIndex({"author": 1, "price": -1})          # creates an index sorted by author (ascending) and then by price (descending).
book_store_A> db.books.find({"author": "Dr. Arif"}).sort("price", -1)    # Such queries will now be faster
```

### (iii) Text Index
- A text index allows full-text search on string fields (like searching words in titles or comments), and is useful when you want to support keyword searches.
- MongoDB’s text index tokenizes and normalizes words to enable keyword searching.
- It supports operators like `$search`, `$language`, and `$caseSensitive`.
```bash

book_store_A> db.reviews.createIndex({"comment": "text", "book": "text"})   # enable keyword search on comments and book titles in reviews collections

- The `find()` method returns an iterable cursor object containing all matching documents, while `find_one()` method returns only the first matching document as a dictionary.
```
db.collection.find()                    # No arguments
db.collection.find(filter)              # One argument (query/filter)

- The  **projection** is a dictionary that specifies which fields to include or exclude

## d. MongoDB Query Language (MQL) - Update Operations  using `pymongo`
- MongoDB uses MQL (MongoDB Query Language), which is a JSON-based query language accessed through API methods like `update_one()`, `update_many()`, and `replace_one()`.
- The `update_one()` method updates the first matching document and returns an UpdateResult object, while `update_many()` method updates all matching documents and returns an UpdateResult object.
```
db.collection.update_one(filter, update)              # Two arguments (filter + update)
db.collection.update_many(filter, update)             # Two arguments (filter + update)
db.collection.update_one(filter, update, upsert=True) # Optional: insert if not found
```

### (i) Update Single Document - `update_one()`

## f. INDEXING
- **What is Indexing?** Indexing in MongoDB improves the efficiency of search operations by allowing the database to locate documents without scanning every one. It's like an index in a textbook.
- **Benefits of Indexing:**
    - Faster queries
    - Improved performance for large datasets
    - Reduced resource usage
    - Indexing also helps in sorting
- **Trade-Offs of Indexing:**
    -  Indexes require additional memory and disk to store.
    -  Insert, update, and delete operations can become slower, as indexes need to be updated each time.
- **When not to use indexing in mongodb:**
    - When the collection is small
    - When the collection is frequently updated
    - When the queries are complex(multiple fields)
    - When the collection is large(make less indices)

- **Types of Indexes in MongoDB:**

| Index Type | Description | Advantage | When to Use | Example |
|------------|-------------|-----------|-------------|---------|
| **Single Field Index** | Creates an index on a single field of a document | Fast queries on one specific field | When filtering or sorting by a single field frequently | `db.users.createIndex({ email: 1 })` - Quick lookups by email |
| **Compound Index** | Creates an index on multiple fields in a specific order | Efficient queries that filter or sort by multiple fields together | When queries use multiple fields in WHERE clauses or sorting | `db.users.createIndex({ lastName: 1, firstName: 1 })` - Search by last name and first name |
| **Text Index** | Creates a full-text search index on string content | Enables searching for words/phrases within text fields | When implementing search functionality across text content | `db.articles.createIndex({ title: "text", content: "text" })` - Search articles by keywords |

### (i) Single Field Index
- A single-field index is created on one field of a collection and is useful when your queries frequently filter or sort by that field.
- Suppose we often search for books by their author. Without an index, MongoDB would scan every document (collection scan), while with the index on author, it can jump directly to matching entries

````python
# creates an ascending index on the author field. 
db.books.create_index({"author": 1})    # 1 = ascending order
# Queries like below will now be faster:
db.books.find({"author": "Dr. Arif"})
````

### (ii) Compound Index
- A compound index includes multiple fields in a single index, and is useful when queries involve filtering or sorting on more than one field.
- Suppose we often search for books by both author and price. Without an index, MongoDB would scan every document (collection scan), while with the index on author and price, it can jump directly to matching entries
- Order matters in compound indexes — {author: 1, price: -1} is different from {price: -1, author: 1}. Design them based on the most common query patterns.

````python
# creates an index sorted by author (ascending) and then by price (descending).
db.books.create_index({"author": 1, "price": -1})
# Queries like below will now be faster:
db.books.find({"author": "Dr. Arif"}).sort("price", -1)
````

### (iii) Text Index
- A text index allows full-text search on string fields (like searching words in titles or comments), and is useful when you want to support keyword searches.
- MongoDB’s text index tokenizes and normalizes words to enable keyword searching.
- It supports operators like `$search`, `$language`, and `$caseSensitive`.

````python
# enable keyword search in book titles or review comments:
# On the books collection
db.books.create_index({"title": "text"})
# On the reviews collection (for multiple fields)
db.reviews.create_index({"comment": "text", "book": "text"})

# Now you can perform full-text searches like:
db.books.find({"$text": {"$search": "MongoDB"}})
db.reviews.find({"$text": {"$search": "Excellent"}})
````

It's used when your data set becomes too large to fit on a single machine or when you need high throughput for reads and writes.</h3>

## Why Use Sharding?

- **Scalability**: Add more machines to handle more data.  
- **Performance**: Queries and writes can be distributed.  
- **Storage Distribution**: Splits large datasets across shards.
