### BERT: The Reader (By Google)
- Bidirectional Encoder Representations from Transformers understands context in both directions, and excels at fill-in-the-blank style tasks, e.g., distinguishing “bank” (financial) from “river bank” (geographical).
- BERT uses the encoder part of the Transformer and is often described as “the reader” because it’s designed to understand and interpret text and performs tasks like classification, sentence similarity and representation learning.
- BERT quickly became base model for **NLU tasks** like question answering, sentiment analysis, and text classification (RoBERTa, DistilBERT, and ALBERT).

<br><br>

### GPT: The Writer (By OpenAI)
- Generative Pre-trained Transformer (GPT) predicts the next word in a sequence, learning to produce coherent and contextually relevant text.
- GPT uses the decoder part of the Transformer and is often called “the writer” because it’s trained to output next-token logits used for autoregressive text generation.
- GPT became the foundation for **NLG tasks** such as story writing, dialogue systems, and code generation, inspiring advanced models like GPT-2, GPT-3, and ChatGPT.

# <span style='background :lightgreen' >8. Sinusoidal Positional Encoding in Transformer Architecture</span>

## a. Why Positional Encoding?

<h3 align="center"><div class="alert alert-success" color=magenta style="margin: 20px">The static token embeddings being fed to transformer DONOT encode word order. </h3>

- **Example** sentences that contain the same words, but their meanings are completely different because the word order changes the subject and object:
    - "The dog bit the man” vs “The man bit the dog”
    - “The chef cooked the fish” vs “The fish cooked the chef”
    - “The police arrested the thief” vs “The thief arrested the police”
    - “The doctor examined the patient” vs “The patient examined the doctor”
    - “I saw the man with a telescope” vs “With a telescope, I saw the man”


<h3 align="center"><div class="alert alert-success" color=magenta style="margin: 20px">Unlike RNNs which process each token separately, the transformer architecture processes all tokens in parallel, so they have no inherent notion of word order.</h3>

<br><br><br><br><br><br><br><br>
<div style="text-align:center;">
    <img src="../images/pe01.png"
         style="max-width:2000px; width:100%; height:auto; display:inline-block;">

## What is Positional Encoding?
- **Solution to above problem is Positional encoding** which gives each token a sense of *first*, *second*, *third*, . . .  position in the sequence.
- Before the input text enters the encoder or decoder blocks of a transformer based models, its tokens are first converted into static embedding vectors.
<br><br><br><br>

<br><br><br><br><br><br><br><br>
## How to Do Positional Encoding?
### **Option 1:** 
- **Add one more dimension in the embedding vector and keep a scalar value there to represent the position of that vector.**

<div style="text-align:center;">
    <img src="../images/pe-02.png"
         style="max-width:2000px; width:100%; height:auto; display:inline-block;">
</div>

### Limitations of Option 1:
- If the input sequence is very long, we will have large values at the position part. Neural networks doesn't work well with large numbers and you may get exploding / valinshing radients during the training phase.

<br><br><br><br><br><br><br><br><br><br><br><br>
### **Option 2:** 
- **To keep the numbers within a bounded range, divide each position value with the total number of tokens.**

<div style="text-align:center;">
    <img src="../images/pe-03.png"
         style="max-width:2000px; width:100%; height:auto; display:inline-block;">
</div>

### Limitations of Option 2:
- If the sequence length is 5 as in this case the second position is encoded as 0.4, but if the sequence length is 10, the second position get encoded as 0.2. So the some token at same position gets a different positional value depending on how long the sentence is.
- There is no sense of relative distance between positions.

<br><br><br><br><br><br><br><br><br><br><br><br>
### **Option 3:** 
- **Use a trigonometric function `sine(pos)`, to compute the position value.**
- The `sin(pos)` produces a float value for ANY position regardless of sequence length and the values always stay between -1 and +1.
- **Limitation** The sine() is periodic, it will repeat same value for certain distant positions, e.g., `sin(0) = sin(π) = sin(2π) = 0.000`. So two completely different positions can produce the EXACT SAME positional encoding value

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(x, np.sin(x), color='#2980b9', linewidth=2)
ax.grid(True, alpha=0.4)

# Build legend text showing sin values for positions 1 to 10 
legend_lines = [f"sin({pos}) = {np.sin(pos):+.4f}" for pos in range(1, 11)]
# Split into two columns of 5 for compact display 
col1 = "\n".join(legend_lines[:5])
col2 = "\n".join(legend_lines[5:])
legend_text = col1 + "\n\n" + col2
ax.text(0.72, 0.97, legend_text,
        transform    = ax.transAxes,
        fontsize     = 9,
        verticalalignment = 'top',
        fontfamily   = 'monospace',
        bbox         = dict(boxstyle='round,pad=0.5',
                            facecolor='#eaf4fb',
                            edgecolor='#2980b9',
                            linewidth=1.2,
                            alpha=0.9))
ax.set_title("Sine Wave  —  sin(x)  for  x ∈ [0, 20]", fontsize=12, fontweight='bold')
ax.set_xlabel("Position (x)", fontsize=10)
ax.set_ylabel("sin(x)",       fontsize=10)
# Mark each integer position 1–10 with a dot on the curve
for pos in range(1, 21):
    ax.plot(pos, np.sin(pos), 'o',
            color='#e74c3c', markersize=6, zorder=5)
    ax.annotate(f"{pos}",
                xy        = (pos, np.sin(pos)),
                xytext    = (pos, np.sin(pos) + 0.15),
                fontsize  = 8,
                ha        = 'center',
                color     = '#e74c3c',
                fontweight= 'bold')
plt.tight_layout()
plt.show()
````

### **Option 4:** 
- **Use two trigonometric functions sin(pos) and cos(pos), and instead of concatenating one additional vlaue, use two values to encode position**
```
  PE(pos) = [sin(pos), cos(pos)]

  position 1 → [sin(1),  cos(1)]  = [ 0.841,  0.540]
  position 2 → [sin(2),  cos(2)]  = [ 0.909, −0.416]
  position 3 → [sin(3),  cos(3)]  = [ 0.141, −0.990]
  position 4 → [sin(4),  cos(4)]  = [−0.757, −0.654]
  position 5 → [sin(5),  cos(5)]  = [−0.959,  0.284]
```

- **Limitation (Position Lives in Only 2 Dimensions:):** 
    - Position is encoded in only 2 out of 512 dimensions, other 510 dimensions carry NO positional information
    - Position should ideally influence ALL 512 dimensions so that every attention calculation is position-aware
    - We need to spread positional information across ALL 512 dimensions of the embedding vector not just 2 extra slots

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 1000)

fig, ax = plt.subplots(figsize=(16, 6))

# ── Plot both curves ──────────────────────────────────────────────────────────
ax.plot(x, np.sin(x), color='#2980b9', linewidth=2, label='sin(pos)')
ax.plot(x, np.cos(x), color='#27ae60', linewidth=2, label='cos(pos)')
ax.grid(True, alpha=0.4)
ax.set_title("Positional Encoding — sin(pos) and cos(pos) for positions 1 to 20",
             fontsize=12, fontweight='bold')
ax.set_xlabel("Position (pos)", fontsize=10)
ax.set_ylabel("Encoding Value", fontsize=10)
ax.set_xlim(0, 20)
ax.legend(fontsize=10, loc='upper center')

# ── Mark positions 1–20 on both curves ───────────────────────────────────────
for pos in range(1, 21):
    ax.plot(pos, np.sin(pos), 'o', color='#2980b9', markersize=5, zorder=5)
    ax.plot(pos, np.cos(pos), 'o', color='#27ae60', markersize=5, zorder=5)

# ── RIGHT TOP legend box — cos values for positions 1–20 ─────────────────────
cos_lines = [f"cos({pos:>2}) = {np.cos(pos):+.4f}" for pos in range(1, 21)]
col1      = "\n".join(cos_lines[:10])
col2      = "\n".join(cos_lines[10:])
cos_text  = col1 + "     " + col2
ax.text(1.01, 0.98, cos_text,
        transform           = ax.transAxes,
        fontsize            = 7.2,
        verticalalignment   = 'top',
        horizontalalignment = 'left',
        fontfamily          = 'monospace',
        bbox                = dict(boxstyle='round,pad=0.5',
                                   facecolor='#eafaf1',
                                   edgecolor='#27ae60',
                                   linewidth=1.2,
                                   alpha=0.92))

# ── RIGHT BOTTOM legend box — sin values for positions 1–20 ──────────────────
sin_lines = [f"sin({pos:>2}) = {np.sin(pos):+.4f}" for pos in range(1, 21)]
col1      = "\n".join(sin_lines[:10])
col2      = "\n".join(sin_lines[10:])
sin_text  = col1 + "     " + col2
ax.text(1.01, 0.02, sin_text,
        transform           = ax.transAxes,
        fontsize            = 7.2,
        verticalalignment   = 'bottom',
        horizontalalignment = 'left',
        fontfamily          = 'monospace',
        bbox                = dict(boxstyle='round,pad=0.5',
                                   facecolor='#eaf4fb',
                                   edgecolor='#2980b9',
                                   linewidth=1.2,
                                   alpha=0.92))

plt.tight_layout()
plt.show()
````

## Option Used by the Authors of Transformer Paper **Multi-Frequency Sinusoidal Encoding**

<h3 align="center"><div class="alert alert-success" color=magenta style="margin: 20px">To inject positional information a <b>Positional Encoding Vector</b> is added to each token embedding vector having the same dimension, and the resulting embedding then will hold <b>semantic + positional</b> information of each token.</h3>

<img align=right src="../images/positionalembeddings.png" width="900">

- The original Transformer paper uses fixed sinusoidal functions to encode positions having following formula (values are precomputed and reused):
- **For even dimensions:** $PE(pos, 2i) = \sin\left(\frac{pos}{10000^{\frac{2i}{d_{model}}}}\right)$       
- **For odd dimensions:** $PE(pos, 2i+1) = \cos\left(\frac{pos}{10000^{\frac{2i}{d_{model}}}}\right)$     

-  where:
    - `pos`: Position of the token in the sequence (0, 1, 2, ...)
    - `i`: Dimension index (0, 1, 2, ..., 255)
    - `d_model`: Model dimension (e.g., 512)
- These  positional encodings are added to token embeddings to inject information about token positions in the sequence, as shown in the figure
>- The constant 10000 is just an arbitrary large base, that ensures that very long sequences can also be handled.
>- Small values of i, will generate high frequency, thus will captures fine grained, local differences between near by tokens
>- Large values of i, will generate low frequency, this will capture coarse, long range infromation useful for distant token relationship




<h3 align="center"><div class="alert alert-success" color=magenta style="margin: 20px">The Positional Token Embeddings now capture both the <b>semantic</b> as well as <b>positional information </b> of each token.</h3>

<img align=right src="../images/pos-encoding.png" width="1000">


## Heatmap of Sinusoidal Positional Encoding
```
The plot shows how sinusoidal positional encoding assigns each token position a unique pattern of sine and cosine values across dimensions,
allowing the Transformer to infer both absolute positions and relative distances between tokens.

````python
import numpy as np
import matplotlib.pyplot as plt

def sinusoidal_positional_encoding(max_position, d_model):
    position = np.arange(max_position)[:, np.newaxis]
    # The original formula pos / 10000^(2i/d_model) is equivalent to pos * (1 / 10000^(2i/d_model)).
    # I use the below version for numerical stability
    div_term = np.exp(np.arange(0, d_model, 2) * -(np.log(10000.0) / d_model))
    pe = np.zeros((max_position, d_model))
    pe[:, 0::2] = np.sin(position * div_term)
    pe[:, 1::2] = np.cos(position * div_term)
    return pe

max_position = 100  # Maximum sequence length
d_model = 128        # Embedding dimension
pe = sinusoidal_positional_encoding(max_position, d_model)

plt.figure(figsize=(12, 8))
plt.imshow(pe, cmap='coolwarm', aspect='auto', vmin=-1, vmax=1)
plt.colorbar()
plt.title('Sinusoidal Positional Encoding')
plt.xlabel('Dimension')
plt.ylabel('Position')
plt.tight_layout()
plt.show()
````

```

PE vectors (calculate as done in the previous example:

pos=1: [0.8415, 0.5403,  0.0100, 0.9999]
